## Step 1: Initial Setup and Company Information

To begin our analysis, we need to install the `yfinance` library, which provides access to historical market data, financial statements, and company information from Yahoo Finance. We'll then select our target company, Duke Energy (DUK), and retrieve its basic information such as sector and market capitalization. This initial data helps us understand the company's basic characteristics and market presence.

```python
# Install the financial data library
!pip install yfinance

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Let's pick a company: Duke Energy (Ticker: DUK)
ticker = "DUK"
company = yf.Ticker(ticker)

# Get basic info to see if it's a "Traditional" or "Alternative" investment
info = company.info
print(f"Company: {info['longName']}")
print(f"Sector: {info['sector']}")
print(f"Market Cap: {info['marketCap']}")
```

### Interpretation of Initial Company Information

From the retrieved information, we can see that Duke Energy is a large-cap company (`Market Cap: 94837882880` or approximately $94.8 billion) operating in the `Utilities` sector. The utilities sector is typically characterized by stable demand, regulated markets, and significant capital expenditures. This initial classification suggests DUK is likely a 'Traditional Investment' rather than an 'Alternative Investment', fitting into the equity markets.

## Step 2: Financial Analysis - Balance Sheet and Debt-to-Equity Ratio

To further understand Duke Energy's financial structure, we will extract key financial data from its balance sheet. Specifically, we aim to:

1.  **Identify the sources of financing**: Determine how much of the company's assets are financed by debt (liabilities) versus equity (ownership capital).
2.  **Calculate the Debt-to-Equity Ratio**: This ratio will tell us the relative proportion of shareholders' equity and debt used to finance a company's assets. A higher ratio indicates more reliance on debt.

```python
# 1. Fetch the Financial Statements (Balance Sheet)
bs = company.balance_sheet

# 2. Extract key metrics for our classification
# We look for Total Liabilities vs Total Stockholder Equity
# This helps us see if they rely on debt (borrowing) or equity (ownership)
total_liabilities = bs.loc['Total Liabilities Net Minority Interest'].iloc[0]
total_equity = bs.loc['Stockholders Equity'].iloc[0]

print(f"Total Liabilities: {total_liabilities:,.0f}")
print(f"Total Equity: {total_equity:,.0f}")
print(f"Debt-to-Equity Ratio: {total_liabilities / total_equity:.2f}")
```

### Interpretation of Capital Structure and Debt-to-Equity Ratio

Our analysis reveals the following for Duke Energy:

*   **Total Liabilities: $142,717,000,000**
*   **Total Equity: $51,842,000,000**
*   **Debt-to-Equity Ratio: 2.75**

**Interpretation:**

1.  **Significant Debt Financing**: The `total_liabilities` are considerably higher than `total_equity`, resulting in a high Debt-to-Equity ratio of 2.75. This indicates that Duke Energy heavily relies on borrowed funds to finance its operations and assets.

2.  **Industry Context**: A high Debt-to-Equity ratio is common for utility companies like Duke Energy. Utilities typically have stable and predictable cash flows, allowing them to comfortably service large amounts of debt. They often undertake massive infrastructure projects that require substantial long-term capital, making debt financing a common and often cost-effective choice.

3.  **Risk and Cost of Capital**: While debt can be a cheaper source of capital due to tax deductibility of interest, a very high ratio can signal increased financial risk if the company's ability to generate cash flow deteriorates. However, given the regulated and essential nature of utilities, their risk profiles are generally stable.

## Step 3: Stock Price Visualization and Market Behavior

Visualizing the stock price history for the last year helps us observe market behavior, liquidity, and overall trend. This is crucial for understanding a company's presence in the public market and how its stock is perceived by investors.

```python
# 3. Visualize the stock price to check for liquidity/market behavior
hist = company.history(period="1y")
hist['Close'].plot(title=f"{ticker} Stock Price (Last 1 Year)", figsize=(10, 5))
plt.ylabel("Price (USD)")
plt.show()
```

### Interpreting the Stock Price Chart for Duke Energy (DUK)

The plot above displays Duke Energy's (DUK) closing stock price over the last year. Here's what we can observe and interpret from this visualization:

DUK exhibits a trend profile consistent with its classification as a utility-sector traditional investment. The chart reveals key support near $112  and   resistance near $132. While the stock maintained an overall positive trend over the last year, the volatility observed in early 2026 is noteworthy. Given the company's Debt-to-Equity ratio of 2.75, this price correction likely reflects market sensitivity to interest rate fluctuations, which remain a primary variable in the valuation of capital-intensive utility assets.

## Step 4: Calculate Annual Percentage Change in Stock Price

To quantify the stock's performance over the last year, we will calculate the annual percentage change. This provides a clear metric for its appreciation or depreciation during the observed period.

```python
first_price = hist['Close'].iloc[0]
last_price = hist['Close'].iloc[-1]
annual_change = ((last_price - first_price) / first_price) * 100
print(f"Annual Percentage Change in Stock Price: {annual_change:.2f}%")
```

### Interpretation of Annual Percentage Change

The annual percentage change in Duke Energy's stock price over the last year is **7.07%**. This indicates a positive return for shareholders during this period, reflecting the market's ongoing valuation of its utility infrastructure business. This appreciation, combined with the stock's relatively stable nature, reinforces its classification as a traditional investment.
